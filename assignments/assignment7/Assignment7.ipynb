The mean in this case is the cumulative final grade the student gets which may be a weighted average of all their assignments. However, a approximately normal distribution on G3 scores in both datasets could also be due to the fact that the schools in the dataset curve and assign grades such that they follow a roughly normal distribution which is common in certain schools.

```python
# Creates a new plot which will show both datasets side by side
plt.figure(figsize=(20,6))
# The first boxplot for Math dataset
plt.subplot(1, 2, 1)
sns.boxplot(x='freetime', y='G3', data=dfm)
plt.title("Math Dataset")
# The second boxplot for Portuguese dataset
plt.subplot(1, 2, 2)
sns.boxplot(x='freetime', y='G3', data=dfp)
plt.title("Portuguese Dataset")

plt.suptitle("Box plots for G3 scores based off of amount of self-reported freetime(0-5 where 5 is most freetime) datasets")
plt.show()
```

##Testing method
To test this I used boxplots to show the distribution of G3 scores among the students based off of the self-reported amount of freetime ranging from 0-5. I used both datasets to verify this for the two different courses we have data on which are Portuguese and Math.

##Results of the hypothesis testing
From the boxplot, we can see that generally as freetime increased, the 25th percentile, median, and 75th percentile G3 scores were higher for the Math dataset but in the Portuguese dataset, the relationship seems to be that as freetime increased, the 25th percentile, median, and 75th percentile G3 scores were lower for the Math dataset.

##Conclusion
