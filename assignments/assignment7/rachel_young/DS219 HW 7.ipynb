## 4. Hypothesis Testing 

### Hypothesis 1: There is a correlation between failures and final grade 

```python
#create box plot to graph distribution of final grades
sns.boxplot(x='failures', y='G3', data=df)
plt.title('Math: Box Plot of Past Failures and Final Grade')
plt.xlabel('Past Failures')
plt.ylabel('Final Grade')
```

```python
#Create pivot table with descriptive statistics
fail = df.pivot_table(index=['failures'], values=['G3'],
                       aggfunc={'median', 'mean', 'min', 'count'})
fail
```

**Interpretation**

From the boxplot, we can see that for final grades in math the more past failures you have, the lower the median final grade is. For example, the median score for students with no failures is 11, but the median scores for those with past failures are descending values between 9 and 7. However, it can be seen that the interquartile range for higher number of failures (3 or 4) is wider and shows greater variability, which means that there is still a wide range of grades that students with 3 or 4 failures can have from about 9 to nearly 0. However, based on the trend in the medians I would say there is a correlation between the two variables.


### Hypothesis 2: There is a correlation between mother's education and final grade 

```python
#create box plot to graph distribution of final grades
sns.boxplot(x='Medu', y='G3', data=df)
plt.title("Math: Box Plot of Mother's Education and Final Grade")
plt.xlabel("Mother's Education")
plt.ylabel('Final Grade')
```

```python
#Create pivot table with descriptive statistics
medu = df.pivot_table(index=['Medu'], values=['G3'],
                       aggfunc={'median', 'mean', 'min', 'count'})
medu
```

**Interpretation**

From the boxplot we can see that students with mother's with the highest level of education (4 - higher education) have a higher median final grade at 12, but there is less of a pattern for the median grades of students with mother's who have education levels lower than that. As seen in the pivot table, for mothers with levels of 1,2, and 3 of education, the median values are about the same at 10, 11, and 10, which breaks the pattern of a positive relationship between mother's education and final grade. Also there are only 3 students with mother's who have no formal education, which can represent bias in our sample where individuals whose parents have low level of education may not have the resources necessary to be well represented in the two schools sampled from. This proves my hypothesis wrong, but there may be a relationship between final grades and whether student's mothers attended higher education specifically. There may be less of a pattenrn of differences in final grades for students with mothers that have education levels lower than higher education.
